```python
#bài 4
a=int(input("nhập chiều cao: "))
for i in range(1,a+1):
    if i==1:
        print(" "*(a-i)+"*")
    elif i>1 and i<a:
        print(" "*(a-i)+"*"+" "*(2*i-3)+"*")
    elif i==a:
        print("*"*(2*a-1))
```

```python
#bài 16
n=int(input("nhập số nguyên dương: "))
ssc=0
ssl=0
for i in range(n+1):
    a=(n//(10**i))%10
    if a==0 and n//(10**i)==0:
        break
    if a%2==0:
        ssc=ssc+1
    else:
        ssl=ssl+1
print(f"Số {n} có {ssc} chữ số chẵn và {ssl} chữ số lẻ")
```

```python
#bài 20
n=int(input("nhập số nguyên dương: "))
so=0
for i in range(n+1):
    a=(n//(10**i))%10
    if a==0 and n//(10**i)==0:
        break
    if a>1:
        for b in range(2,a):
            if a%b==0:
                break
        else:
            so=so+1
print(so)        
   
```
