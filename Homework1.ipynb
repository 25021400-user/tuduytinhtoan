```python
#bài 6
a,b,c = map(float, input("Nhập 3 số thực: ").split())
if a+b>c and a+c>b and b+c>a:
    print("3 số này có thể là độ dài 3 cạnh của một tam giác")
    p=(a+b+c)/2
    s=(p*(p-a)*(p-b)*(p-c))**0.5
    print(f"Diện tích của tam giác là: {s:.1f}")
else:
    print("3 số này không thể là độ dài 3 cạnh của một tam giác")
```

```python
a=input("tên chủ hộ:")
b=int(input("chỉ số tháng trước"))
c=int(input("chỉ số tháng này"))
d=c-b
print(f"Ho va ten:{a}")
if d <= 50:
    T=d*1984*(1.08)
    print(f" Tien phai tra la: {T:.0f} ")
elif d <= 100 and d > 50:
    T=50*1984*(1.08)+(d-50)*2050*(1.08)
    print(f" Tien phai tra la: {T:.0f} ")
elif d <= 200 and d > 100:
    T=50*1984*(1.08)+50*2050*(1.08)+(d-100)*2380*(1.08)
    print(f" Tien phai tra la: {T:.0f} ")
elif d <= 300 and d > 200:
    T=50*1984*(1.08)+50*2050*(1.08)+100*2380*(1.08)+(d-200)*2998*(1.08)
    print(f" Tien phai tra la: {T:.0f} ")
elif d > 300 and d <= 400:
    T=50*1984*(1.08)+50*2050*(1.08)+100*2380*(1.08)+100*2998*(1.08)+(d-300)*3350*(1.08)
    print(f" Tien phai tra la: {T:.0f} ")
elif d > 400:
    T=50*1984*(1.08)+50*2050*(1.08)+100*2380*(1.08)+100*2998*(1.08)+100*3350*(1.08)+(d-400)*3460*(1.08)
    print(f" Tien phai tra la: {T:.0f} ")
```
