```python
#w5a8
def nhiphan(a):
    sonhiphan=""
    while a>0:
        c=a%2
        sonhiphan=str(c) + sonhiphan
        a=a//2
    return sonhiphan
def hamming(a,b):
    x=len(nhiphan(a))
    y=len(nhiphan(b))
    c=abs(x-y)
    return c
a,b= map(int,input('nhap 2 so nguyen duong:').split())
print(f'khoang cach hamming giua {a} va {b} la:{hamming(a,b)}')
```

```python
#w5a15
def is_right_triangle(a,b,c):
    if a+b>c and a+c>b and b+c>a:
        if a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
            return True
        else:
            return False
    else:
        return False
a,b,c= map(float, input("Nhập 3 cạnh của tam giác: ").split())
print(is_right_triangle(a,b,c))
```
