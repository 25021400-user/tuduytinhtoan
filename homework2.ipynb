```python
#bài 3
a,b,c= map(float, input("Nhập 3 cạnh của tam giác: ").split())
p=(a+b+c)/2
s=(p*(p-a)*(p-b)*(p-c))**0.5
if a+b>c and a+c>b and b+c>a:
    if a==b==c:
        print("Tam giác đều")
    elif a==b or a==c or b==c:
        print("Tam giác cân")
    elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
        print("Tam giác vuông")
    else:
        print("Tam giác thường")
    print(f"Diện tích tam giác là: {s}")
    print(f"Chu vi tam giác là: {p}")
else:
    print("không là ba cạnh của một tam giác.")
    
```

```python
#bài 5
a,b,c=map(float, input("Nhập 3 hệ số a,b,c: ").split())
delta=b**2-4*a*c
if a==0:
    print("không phải phương trình bậc 2")
elif delta<0:
    print("phương trình vô nghiệm")
elif delta==0:
    x=-b/(2*a)
    print(f"phương trình có nghiệm kép x={x}")
else:
    x1=(-b+delta**0.5)/(2*a)
    x2=(-b-delta**0.5)/(2*a)
    print(f"phương trình có hai nghiệm phân biệt x1={x1} và x2={x2}")
```

```python
#bài 8
a,b,c,d,m,n= map(float, input("Nhập các hệ số: ").split())
det=a*d-c*b
if det==0:  
    if a*n==c*m and b*n==d*m:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
else:
    x=(d*m-c*n)/det
    y=(-b*m+a*n)/det
    print(f"Hệ phương trình có nghiệm duy nhất: x={x}, y={y}")
```

```python
#bài 10
a,b= map(float, input("Nhập tọa  của tâm độ (x y): ").split())
c=float(input("Nhập bán kính hình tròn: "))
d,e= map(float, input("Nhập tọa độ điểm cần kiểm tra (x y): ").split())
distance=((d-a)**2+(e-b)**2)**0.5
if distance<c:
    print("Điểm nằm trong hình tròn")
elif distance==c:
    print("Điểm nằm trên đường tròn")
else:
    print("Điểm nằm ngoài hình tròn")
```
